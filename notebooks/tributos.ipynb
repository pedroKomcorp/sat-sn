# PIS COFINS

```python
import pandas as pd

icms_path = '../resources/icms.xlsx'
piscofins_path = '../resources/piscofins.xlsx'
```

```python
df_piscofins = pd.read_excel(piscofins_path, engine='calamine')

df_piscofins['NCM'] = df_piscofins['NCM'].str.replace('.', '', regex=False)

df_piscofins_updated = df_piscofins
df_piscofins_updated = df_piscofins_updated.drop(['Obs: Planilha exportada do ITC em 11/06/2024'], axis=1)
```

```python

df_sieg = pd.read_excel('../resources/sieg/sieg.xlsx', sheet_name='PRODUTOS', dtype={'NCM': str}, engine='calamine')
```

```python
# --- ETAPA DE LIMPEZA ---
# Garante que ambas as colunas NCM são do tipo string (texto)
df_sieg['NCM'] = df_sieg['NCM'].astype(str)
df_piscofins_updated['NCM'] = df_piscofins_updated['NCM'].astype(str)

# Remove espaços em branco do início e do fim das colunas
df_sieg['NCM'] = df_sieg['NCM'].str.strip()
df_piscofins_updated['NCM'] = df_piscofins_updated['NCM'].str.strip()


# --- SEU CÓDIGO ORIGINAL (AGORA MAIS SEGURO) ---

# 1. Crie o mapa de consulta (lookup map) a partir do DataFrame já limpo
piscofins_indexed = df_piscofins_updated.set_index('NCM')['TRIBUTAÇÃO']

# 2. Mapeie os valores para a coluna 'PIS COFINS'
df_sieg['PIS COFINS'] = df_sieg['NCM'].map(piscofins_indexed)
df_sieg['PIS COFINS'] = df_sieg['PIS COFINS'].fillna('NCM NÃO IDENTIFICADO')

df_sieg = df_sieg[['PIS COFINS'] + [col for col in df_sieg.columns if col != 'PIS COFINS']]
df_sieg

```

# ICMS

```python
df_icms = pd.read_excel(icms_path, engine='calamine', dtype={
  'INICIO': str, 'FIM': str, 'NCM': str
})
df_icms['INICIO'] = pd.to_datetime(df_icms['INICIO'], dayfirst=True, format='mixed')
df_icms['FIM'] = pd.to_datetime(df_icms['FIM'], dayfirst=True, format='%d/%m/%Y')


df_icms['INICIO'] = df_icms['INICIO'].dt.strftime('%d/%m/%Y')
df_icms['FIM'] = df_icms['FIM'].dt.strftime('%d/%m/%Y')


df_icms
```

```python
df_sieg['DATA EMISSAO'] = pd.to_datetime(df_sieg['DATA EMISSAO'], format='%Y-%m-%d')
df_sieg['DATA EMISSAO'] = pd.to_datetime(df_sieg['DATA EMISSAO'].dt.strftime('%d/%m/%Y'), dayfirst=True)
df_sieg
```

```python
df_icms['INICIO'] = df_icms['INICIO'].fillna(pd.to_datetime('01/01/1990', dayfirst=True))
df_icms['FIM'] = df_icms['FIM'].fillna(pd.to_datetime('31/12/2050', dayfirst=True))
```

```python
merged = df_sieg.merge(df_icms, on="NCM", how="left", suffixes=("", "_icms"))

# Ensure dates are datetime
merged["DATA EMISSAO"] = pd.to_datetime(merged["DATA EMISSAO"])
merged["INICIO"] = pd.to_datetime(merged["INICIO"])
merged["FIM"] = pd.to_datetime(merged["FIM"])

def classify(row):
    if pd.isna(row["INICIO"]):  
        return 'NCM não encontrado'  # No NCM found
    if row["INICIO"] <= row["DATA EMISSAO"] <= row["FIM"]:
        return row["CLASSIFICAÇÃO"]
    return "NÃO ST"

merged["ICMS"] = merged.apply(classify, axis=1)

# Keep only original SIEG columns + ICMS
final = merged[df_sieg.columns.tolist() + ["ICMS"]]
```

```python
final = final[['ICMS'] + [col for col in final.columns if col != 'ICMS']]
final.to_csv('OUTPUT.CSV')
```
